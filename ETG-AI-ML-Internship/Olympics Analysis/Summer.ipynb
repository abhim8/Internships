# Summer Olympics Data Analysis Assignment

```python
# initial code to import libraries and read data-set
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("/content/summer.csv")
```

### 1. In how many cities Summer Olympics is held so far?

```python
data = []
no_of_cities = 0

for city in df['City'].unique():
    data.append([city , len(df[df['City']  == city])])
    no_of_cities = no_of_cities + 1

# total no of cities
print('Total no of Cities = '+ str(no_of_cities))

# bar graph for all the available cities, based on frequency occured 
pd.DataFrame(data,columns = ['City','No Of Games']).sort_values(by='No Of Games', ascending=False).plot(x = 'City', y = 'No Of Games', kind = 'bar', figsize = (no_of_cities,5))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
data = []
for g in df['Sport'].unique():
    data.append([g , len(df[df['Sport']  == g])])

pd.DataFrame(data,columns = ['Sport','Gold Medals']).sort_values(by='Gold Medals', ascending=False).head().plot(x = 'Sport', y = 'Gold Medals', kind = 'bar', figsize = (5,5))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data = []

for sp in df['Sport'].unique():
    data.append([sp , len(df[df['Sport']  == sp])])

pd.DataFrame(data,columns = ['Sport','No Of Medals']).sort_values(by='No Of Medals', ascending=False).head().plot(x = 'Sport', y = 'No Of Medals', kind = 'bar', figsize = (5,6))
```

### 4. Which player has won most number of medals? (Top 5)

```python
data = []

for ath in df['Athlete'].unique():
    data.append([ath , len(df[df['Athlete']  == ath])])

pd.DataFrame(data,columns = ['Athlete','No Of Medals']).sort_values(by='No Of Medals', ascending=False).head().plot(x = 'Athlete', y = 'No Of Medals', kind = 'bar', figsize = (5,5))
```
